7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar a função map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é classificado como promotor se a média das notas é maior ou igual a 80, como neutro se a média estiver entre 50 e 80, e como detrator se for abaixo de 50

```python
df_total['Total'] = df_total[['Ativo01','Ativo02','Ativo03','Ativo04','Ativo05','Ativo06','Ativo07']].sum(axis=1)
df_total['Media_notas'] = df_total[['Nota1', 'Nota2']].mean(axis=1)
df_total['Dominio'] = df_total['Email'].str.split('@').str[1]

df_total.rename(columns={'Salario': 'Renda'}, inplace=True)
df_total['Renda'] = df_total['Renda'].str.replace('$','', regex=False).str.replace(',','').astype(float)

df_total['Sexo'] = df_total['Sexo'].map({'Masc': 'Masculino', 'Fem': 'Feminino'})

df_total['Media_notas'] = df_total['Media_notas'].fillna(0).astype(int)

condicoes = [
    (df_total['Media_notas'] >= 80),
    (df_total['Media_notas'] >= 50) & (df_total['Media_notas'] < 80),
    (df_total['Media_notas'] < 50)
]
classificacao = ['Promotor', 'Neutro', 'Detrator']
df_total['NPS'] = np.select(condicoes, classificacao)

df_total.head()
```
