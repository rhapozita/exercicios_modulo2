# Numpy: Numeric computing library

NumPy (Numerical Python) é uma das bibliotecas core para computação numérica no Python. Bibliotecas como o Pandas, Matplotlib, Statmodels entre outras bibliotecas cientificas utilizam do seu poder.

As maiores vantagens do Numpy são:
* Computação numérica com os primitivos do C
* Operações vetorizadas
* Uma API de Algebra Linear integrada
* Um API em C para conectar o NumPy com bibliotecas escritas em C, C++ or FORTRAN

1D array = unidimensional - armazena uma lista de variáveis/elementos do mesmo tipo e pode acessar cada var/elemento usando o index

2D array = bi dimensional - armazena dados com formato consistente de linhas e colunas

Pq usar arrays em vez de listas?

- Arrays usam menos memória
- Arrays tem muito mais funcionalidades
- Arrays obrigam os dados serem homogêneos
- Aritmética de arrays operam como manipulação de matriz

2D

- 0 é linha e 1 coluna para array de 2 D
- Atenção: você percorre o eixo, é através do eixo
- axis é o eixo de agragação

```python
b = np.array([[1,-2,3], [4,5,-6]])
```

```python
 #numpy.sum(a, axis=None, dtype=None, out=None, keepdims=<no value>, initial=<no value>, where=<no value>)
# Sum of array elements over a given axis.
np.sum(b)
```

```python
b
```

```python
# somando por meio do eixo 0, a linha
np.sum(b, axis=0)
```

```python
# somando  1+(-2)+3 = 2 (que é a linha 0) e 4+5+(-6)=3 (que é a linha 1) estou agregando(somando) pelo eixo1 
# somando todos os elementos da linha 0 e linha 1
np.sum(b, axis=1)
```

```python
# funciona min, max, mean, cumsum, std, sort para 2D
# ordenar o array b pelo eixo 1
np.min(b,axis=1)
```
